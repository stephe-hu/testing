```python
import geopandas as gpd
import numpy as np
```

# Dataset #1

```python
df_cancer = gpd.read_file('https://raw.githubusercontent.com/stephe-hu/datasci_7_geospatial/main/datasets/Cancer_Rates.geojson')
df_cancer
```

```python
# Identify missing values
missing_values = df_cancer.isnull().sum()
missing_values
```

```python
df_cancer[['All_Cancer', 'ZIP']].sort_values(by=['All_Cancer'], ascending=False)
```

```python
df_cancer['All_Cancer'] = (df_cancer['All_Cancer']/1000)
```

```python
df_cancer.plot('All_Cancer', legend=True)
```

## Analysis Insights
### Description: The dataset measures the cancer rates for Lake County Illinois and provides a rate per 100,000 for each cancer type.
### Anomolies: The lower right region has the highest rate of all cancer (4.5%).

# Dataset #2

```python
df_obesity = gpd.read_file('https://raw.githubusercontent.com/stephe-hu/datasci_7_geospatial/main/datasets/National_Obesity_By_State.geojson')
df_obesity
```

```python
# Identify missing values
missing_values = df_obesity.isnull().sum()
missing_values
```

```python
# Drop rows with any NaN values
df_obesity.dropna(inplace=True)
df_obesity
```

```python
df_obesity[['Obesity', 'NAME']].sort_values(by=['Obesity'], ascending=False)
```

```python
df_obesity.plot('Obesity', legend=True)
```

## Analysis Insights
### Description: The dataset measures the national obesity rate for each state as of 2015.
### Anomolies: The states with the highest obesity are in the eastern part of the country, with Louisiana having the highest obesity rate (36.2%).
